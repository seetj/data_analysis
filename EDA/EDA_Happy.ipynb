```python
import pandas as pd
import numpy as np
import seaborn as sns                       
import matplotlib.pyplot as plt           
%matplotlib inline     
sns.set(color_codes=True)
```

```python
df = pd.read_csv('2015.csv')
df
```

```python
df.dtypes
```

```python
df.describe()
```

```python
df_region = pd.pivot_table(df,index = ['Region'])
df_region
```

```python
df_region['Happiness Rank'].nsmallest(10).plot(kind = 'bar', figsize = (12,5))
plt.xlabel('Region')
plt.ylabel('Happiness Rank')
plt.title('Average Happiness Rank per Region')
```

```python
plt.figure(figsize=(10,5))
c= df_region.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
#According to this heatmap, happiness score is heavily correlated with
#economy, family, freedom, health, and trust
```

```python
#trying out seasborn plot function
plt.figure(figsize= (12,8))
ax = sns.barplot(df['Region'],df['Happiness Rank'])
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
```
